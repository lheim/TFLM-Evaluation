```python
def read_model_information(filename, series):
    if "PRUNED" in filename:
        series['pruned'] = 1
    else:
        series['pruned'] = 0
    if "optimized" in filename:
        series['weights'] = 'int8'
        series['activations'] = 'float32'
    elif "INT8" in filename:
        series['weights'] = 'int8'
        series['activations'] = 'int8'
    else:
        series['weights'] = 'float32'
        series['activations'] = 'float32'
        
    
```

```python
def hbarplot(title, x, x_label, y, x_err=None):
    plt.rcdefaults()
    fig, ax = plt.subplots()

    y_pos = np.arange(len(y))

    ax.barh(y_pos, x, xerr=x_err, align='center')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(y)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(x_label)
    ax.set_title(title)

    plt.savefig(f'{title}.pdf')
    plt.show()
```
