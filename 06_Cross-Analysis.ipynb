```python
hue = 'relative_type'
model_types = df['model_type'].unique()

filter_dict = {
    'optimization': 'int8_none'
}
#filter
#for model_type in model_types:

model_name = 'LeNet'
filter = (df['model'] == model_name)
filter_descr = ''
for key in filter_dict:
    filter = filter & (df[key] == filter_dict[key])
    filter_descr += (str(filter_dict[key]) + '_')
filter = filter & (df['relative_type'] != 'memory') 
 

df_filtered = df.loc[filter]
# df_filtered = df.copy()


fig = sns.catplot(y='layer', x='relative_number', kind='bar', hue=hue, data=df_filtered, legend=True)
#fig = sns.catplot(y='layer', x='relative_number', kind='bar', hue=hue, row='optimization', data=df_filtered, legend=True)
display(df_filtered[['layer', 'optimization', 'relative_type', 'relative_number']])


# plt.title(model_name)

plt.xlabel("Relative Amount of whole NN")
plt.ylabel("Selected Layers")


filename = f"figures/complexity/{model_name}_{filter_dict['optimization']}_complexity_hue-{hue}"

print(filename)

# plt.legend(title='Smoker', loc='upper left', labels=['Hell Yeh', 'Nah Bruh', 'k', 'll'])
# legend = ax.legend()
# legend.texts[0].set_text("Whatever else")

plt.tight_layout()
plt.show()
plt.savefig(filename + '.pdf')
tikzplotlib.save(filename + '.tex', axis_width=AXIS_WIDTH, axis_height='180pt')
```
